```python
n_bytes = 2**31
max_bytes = 2**31 - 1

path = "/Users/maimai/Desktop/project/May/data/givenchy/pickle"  

def write_to_pickle(file_path, data):

    bytes_out = pickle.dumps(data)
    with open(file_path, 'wb') as f_out:
        for idx in range(0, n_bytes, max_bytes):
            f_out.write(bytes_out[idx:idx+max_bytes])

def load_pickle(file_path):
    bytes_in = bytearray(0)
    input_size = os.path.getsize(file_path)
    with open(file_path, 'rb') as f_in:
        for _ in range(0, input_size, max_bytes):
            bytes_in += f_in.read(max_bytes)
    return pickle.loads(bytes_in)

def loadModel(self):
    read_file = "data/" + self.model_name + ".dat"
    self.model = joblib.load(read_file)
    return self.model
```

```python
with plt.style.context(('ggplot')):
    plt.figure(figsize=(15,9))
    x=np.array([20,54,88,122,156,191,225])
#     x=np.linspace(0,150,5)
    single = np.array([20,27,64,325,1932,12568,16907])
    sequential_small_to_large = np.array([5,10,17,49,395,4143,19493])
    sequential_large_to_small = np.array([5,10,31,154,680,6122,26526])
    dynamic = np.array([5,10,47,553,4929,18608,34404])
    
    plt.plot(x,single,color='#D2691E',label='Single-stage Seeding Strategy')
    plt.plot(x,sequential_small_to_large,color='green',label='Sequential_Degree_Small_to_Large')
    plt.plot(x,sequential_large_to_small,color='red',label='Sequential_Degree_Large_to_Small')
    plt.plot(x,dynamic,color='blue',label='Dynamic_Sequential Seeding Strategy')


    plt.xlabel('Time_lapsed (each 30 mins)')
    plt.ylabel('The number of infected nodes')
    plt.legend(loc='upper left')
    plt.title("")
    for a, b in zip(x, single):
        plt.text(a, b, b, ha='right', va='bottom', fontsize=15,color='#D2691E')
    for a, b in zip(x, sequential_small_to_large):
        plt.text(a, b, b, ha='left', va='bottom', fontsize=15,color='green')
    for a, b in zip(x, sequential_large_to_small):
        plt.text(a, b, b, ha='right', va='top', fontsize=15,color='black')
    for a, b in zip(x, dynamic):
        plt.text(a, b, b, ha='left', va='top', fontsize=15,color='blue')
    plt.savefig(path+"/graph/four_sequences.png")
    plt.show()
```
